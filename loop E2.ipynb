#looping through string
for i in "apple":

```python
fruits = ["apple", "orange", "pinapple"]
for f in fruits:
    # print(f) break statement
    if f == "orange":
        break
    print(f)
```

```python
#continue statement
fruits = ["apple", "orange", "pinapple"]
for f in fruits:
    if f == "orange":
        continue
    print(f)
```

#else in for loop


```python
for x in range(6):
    print(x)
else:
    print("finally finished")
```

```python
for x in range(6):
    if x == 3:
        #break
        print(x)
else:
    print("finally finished")
```

```python
for x in range(6):
    if x == 3:
        break
    print(x)
else:
    print("finally finished")
```

#nested loop

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]
for x in adj:
    for y in fruits:
        print(x,y)
```
