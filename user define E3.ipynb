#example of user defined function

```python
def add_numbers(x,y):
    sum = x + y
    return sum
num1 = 10
num2 = 15
print("The sum is" , add_numbers(num1,num2))
```

```python
def add_numbers(x,y):
    sum = x + y
    return sum
num1 = int(input("Enter first number :"))
num2 = int(input("Enter second number :"))
print("The sum is", add_numbers(num1,num2))
```

 #function defination

```python
def find_square(a):
    result = a * a
    return result
a = int(input("Enter number: "))
print("The square of the number is: ",find_square(a))
```

#function call

```python
square = find_square(5)
print("square of the number is:",square)
```
